```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### read in csvs
dir = 'int_outputs/'

files = os.listdir(dir)

DF_M = pd.DataFrame()

for file in files:
    if '.csv' in file:
        df = pd.read_csv(dir+file)
        DF_M = pd.concat([DF_M, df])

### set up a color palette for plots
keys = DF_M['Experiment'].unique()
values = ['r', 'b', 'g', 'purple', 'gray', 'orange', 'pink']
palette = dict(zip(keys, values))
```

```python
var = 'tasmin'

DF = DF_M.dropna(subset=var)
DF_ens_means = DF.groupby(['Experiment', 'year']).mean().reset_index()
DF_ens_means
for exp in DF_ens_means['Experiment'].unique():
    plt_data = DF_ens_means[DF_ens_means['Experiment']==exp]
    plt_data = plt_data[plt_data['year']<2100]
    plt.plot(plt_data['year'], plt_data[var],
             label=exp, color=palette[exp])
    plt.xlabel('year')
    plt.ylabel(var)
    plt.legend()
    plt.title('{}: Ensemble means'.format(var))
    
```

```python
var = 'tasmin'

DF = DF_M.dropna(subset=var)

for exp in DF['Experiment'].unique():
    plt_data_exp = DF[DF['Experiment']==exp]
    i=0
    for ens in plt_data_exp['Ensemble_member'].unique():
        plt_data = plt_data_exp[plt_data_exp['Ensemble_member']==ens]        
        plt_data = plt_data[plt_data['year']<2100]
        plt.plot(plt_data['year'], plt_data[var],
                 color=palette[exp],
                 label=exp if i == 0 else "")
        i=i+1
    plt.xlabel('year')
    plt.ylabel(var)
    plt.legend()
    plt.title('{}: All ensemble members'.format(var))
    
```
